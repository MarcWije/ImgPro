```python
#Question 2 - Gamma Correction

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im1 = cv.imread('images/highlights_and_shadows.jpg')
assert im1 is not None
im1 = cv.cvtColor(im1, cv.COLOR_BGR2RGB)
#part a
gamma= 0.5
invG = 1.0/gamma
conv = np.array([((i/255.0)**invG) * 255 for i in range(256)], dtype=np.uint8) 

im2 = conv[im1]

h = cv.calcHist([im1],[0], None, [256],[0,256])
h2 = cv.calcHist([im2],[0], None, [256],[0,256])

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Displays images
axs[0, 0].imshow(im1, vmin=0, vmax=255)
axs[0, 0].set_title('Original')
axs[0, 1].imshow(im2, vmin=0, vmax=255)
axs[0, 1].set_title("Gamma corrected image , Gamma: {:.2f}".format(gamma))

# Plot histograms 
axs[1, 0].bar(range(256), h.ravel())
axs[1, 0].set_title('Histogram of Original Image')
axs[1, 1].bar(range(256), h2.ravel())
axs[1, 1].set_title("Histogram of Gamma Corrected Image, Gamma: {:.2f}".format(gamma))
plt.tight_layout()
plt.show()
```
