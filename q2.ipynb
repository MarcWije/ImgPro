```python
#Question 2 - Gamma Correction

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im1 = cv.imread('images/highlights_and_shadows.jpg')
assert im1 is not None

#part a
gamma= 0.5
invG = 1.0/gamma
h = np.array([((i/255.0)**invG) * 255 for i in range(256)], dtype=np.uint8) 

im2 = h[im1]

#part b
def hist(image):
    m = np.zeros(256, dtype=np.uint16)
    for r in range(image.shape[0]):
        for c in range(image.shape[1]):
            m[image[r, c]] += 1
    return m

mat = hist(im1)
mat2 = hist(im2)

cv.imshow('Original image', im1)
cv.imshow("Gamma corrected image , Gamma: {:.2f}".format(gamma), im2)
cv.waitKey(0)
cv.destroyAllWindows()


plt.bar(range(256), mat)
plt.title('Histogram of Original Image')
plt.show()
plt.bar(range(256), mat2)
plt.title("Histogram of Gamma Corrected Image, Gamma: {:.2f}".format(gamma))
plt.show()
```
